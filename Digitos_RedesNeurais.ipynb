```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
digits = load_digits()
X, y = digits.data, digits.target


print(f"Formato dos dados (X): {X.shape}")
print(f"Formato dos labels (y): {y.shape}")

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=18)

print("\n--- TREINAMENTO ---")
mlp_padrao = MLPClassifier(random_state=18)

mlp_padrao.fit(X_train, y_train)

y_pred_padrao = mlp_padrao.predict(X_test)
accuracy_padrao = accuracy_score(y_test, y_pred_padrao)

print(f"Acurácia: {accuracy_padrao:.4f}")

print("\n--- TREINAMENTO OTIMIZADO ---")
mlp_otimizado = MLPClassifier(
    hidden_layer_sizes=(100, 50),  # Duas camadas ocultas: uma com 100 neurônios, outra com 50
    activation='relu',             # Função de ativação ReLU (padrão, mas explícito aqui)
    solver='adam',                 # Otimizador Adam (padrão, mas explícito aqui)
    max_iter=500,                  # Aumentar o número máximo de épocas para garantir a convergência
    alpha=0.0001,                  # Parâmetro de regularização para evitar overfitting
    learning_rate_init=0.001,      # Taxa de aprendizado inicial
    random_state=18                # Garantir reprodutibilidade
)

mlp_otimizado.fit(X_train, y_train)

y_pred_otimizado = mlp_otimizado.predict(X_test)
accuracy_otimizado = accuracy_score(y_test, y_pred_otimizado)

print(f"Acurácia do modelo otimizado: {accuracy_otimizado:.4f}")

print("\nRelatório de Classificação (Otimizado):")
print(classification_report(y_test, y_pred_otimizado))

print("\nMatriz de Confusão (Otimizado):")
cm = confusion_matrix(y_test, y_pred_otimizado)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('Dígito Previsto')
plt.ylabel('Dígito Verdadeiro')
plt.title('Matriz de Confusão')
plt.show()
```
